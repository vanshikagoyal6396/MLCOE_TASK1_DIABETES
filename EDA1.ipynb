***FILL MISSING VALUES(0S)***
REPLACE THE 0 VALUES BY MEDIAN VALUES AS 0 VALUES ARE NOT POSSIBLE

*   Glucose
*   BloodPressure
*   SkinThickness
*   Insulin
*   BMI



```python
data_copy=data.copy()     # making a copy of opriginal data
```

```python
data_copy=data['Glucose'].replace(0,data['Glucose'].median())
data_copy=data['BloodPressure'].replace(0,data['BloodPressure'].median())
data_copy=data['SkinThickness'].replace(0,data['SkinThickness'].median())
data_copy=data['Insulin'].replace(0,data['Insulin'].median())
data_copy=data['BMI'].replace(0,data['BMI'].median())
```
