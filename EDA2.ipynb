***ANALYZING THE AVERAGE OF MOST COLOR PURCHASED WITH THE PURCHASED AMOUNT***

```python
r=pd.DataFrame(data.groupby('Color')['Purchase Amount (USD)'].mean().reset_index().sort_values(by='Purchase Amount (USD)',ascending=False))
r
```

***COUNTPLOT WHICH SHOWS THE COLORS AVAILABLE ***

```python
sns.countplot(x='Color',data=data)
plt.xticks(rotation=90)
plt.show()
```

1. THIS GRAPH SHOWS THAT THE MOST PURCHASED COLOR IS GREEN.
2. THE LEAST PURCHASED COOR IS GOLD.

***RELATION BETWEEN AVAERAGE COLOR PURCHASED THE MOST**

```python
sns.barplot(x='Color',y='Purchase Amount (USD)',data=r)
plt.xticks(rotation=90)
plt.show()

```

***ANALYZING RELATION BETWEEN SEASON,GENDER AND THE AVERAGE PURCHASE AMOUNT***

```python
season=pd.DataFrame(data.groupby('Season')['Purchase Amount (USD)'].mean().reset_index().sort_values(by='Purchase Amount (USD)',ascending=False))
season
```

```python
sns.catplot(x='Season',y='Purchase Amount (USD)',kind='bar',hue='Gender',data=data)
plt.xticks(rotation=90)
plt.show()
```

1.This graph show that maximum average purchase is in Fall season .
2. Minimun purchase is the Summer season.
3. There is uniform distribution  between the male and female

***ANALYZING RELATION BETWEEN CATEGORY ANS THE PURCHASED AMOUNT(USD)***

```python
cat=pd.DataFrame(data.groupby('Category')['Purchase Amount (USD)'].mean().reset_index().sort_values(by='Purchase Amount (USD)',ascending=False))
cat
```

```python
sns.boxplot(x='Category',y='Purchase Amount (USD)',hue='Gender',data=data)
plt.xticks(rotation=90)
plt.show()
```

 We analyse that
 1. Footwear has the maximum average purchase
 2. Outerwear has the minimum average purchase
 3. There is almost the uniform distribution of purchased amound in between the gender

***ANALYSING THE RELATION BETWEEN THE SIZE AND PURHASED AMOUNT ***

```python
size=pd.DataFrame(data.groupby('Size')['Purchase Amount (USD)'].mean().reset_index().sort_values(by='Purchase Amount (USD)',ascending=False))
size
```

```python
sns.countplot(x='Size',data=data)
```

```python
sns.barplot(x='Size',y='Purchase Amount (USD)',data=size)
```

We analyse that
1. The count of MEDIUM  size is the most and the count of XL is minimum
2. But the average Purchase of SMALL  size is the maximum
3. There is small difference between the S,L,XL,M average purchase amount.
